### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:
* $ div = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= \frac{\partial (U+c_1x+c_2y)}{\partial x} + \frac{\partial (V+c_3x+c_4y)}{\partial y} = c_1+c_4$
* $ \zeta_{rel} = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y} = \frac{\partial (V+c_3x+c_4y)}{\partial x} - \frac{\partial (U+c_1x+c_2y)}{\partial y} =  c_3 - c_2$
* $ def1 = \frac{\partial u}{\partial x} - \frac{\partial v}{\partial y} = \frac{\partial (U+c_1x+c_2y)}{\partial x} - \frac{\partial (V+c_3x+c_4y)}{\partial y}= c_1 - c_4$
* $ def2 = \frac{\partial v}{\partial x} + \frac{\partial u}{\partial y} =\frac{\partial (V+c_3x+c_4y)}{\partial x} + \frac{\partial (U+c_1x+c_2y)}{\partial y} = c_3 + c_2$

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

In the case of vorticity plus Def1, the flow is sheared along an axis that appears to be y = x. The further you move away from this axis, the stronger the flow (i.e. longer wind vectors). Notice that here too the wind vectors run parallel to a vertical line at x = 0.

## Answer to Exercise 4: 

Vorticity plus straight flow can be described as ... simply an incipient vortex with adjustments to the net flow due to the background wind. That is, as the "flow" gets stronger, then the center of the vortex seemingly shifts upward. This is simply a result of the impact of adding a background flow to pure vorticity originally located with a center at x = 0 and y = 0. In the original vortex, the further away you move in a radius from (0,0) the stronger the magnitudes of wind vectors. 

Adding some background flow of positive u (or, flow entirely in the x direciton) will shift the center of the vortex up to whatever the strength of U. For example, in our U=4 vortex, the center of the rotation shifts upwards to y = 4, x = 0. This is simply due to the addition of the background flow to existing vortex flow: adding u = 4 to a wind vector at (0,0) will change increase the vector magntitude positively. 

This can easily be related to the atmosphere as represented by flow on a cartesian x-y plane. In the real atmosphere, the U component typically increases due to the thermal wind balance (relating wind shear to horizontal temperature gradients). So, a <b>vortex at the following levels of the atmosphere:</b>

<li>At lower levels in the atmosphere, U is smaller than at higher levels. Thus the center of the vortex, which could represent a low pressure/height surface in the atmosphere, is further "south". However, moving up in the atmosphere (analogous to increasing U like I did in my plots) will slowly shift the center of the vortex and low-pressure center further north and increase the flow around the southern fringe of such a system. Over our domain, the curved flow over the southern domain will increase. (Plots with U = 2 and 4)
</li>
<li>At upper levels, the U will become so strong that the the low pressure/height surface shifts significantly "north" from further down in the atmosphere, to the point that this looks more like a "trough" than a vortex. this is analogous to what is often seen in mid-latitude systems, with the upper-level low pressure/height surface located north of the low-level center. (Plots with U = 8 and U = 16)
</li>

Ultimately, these plots look somewhat similar to a typical mid-latitude extratropical cyclone, with pressure surfaces and the center of the related rotational field at smaller U flows (lower in atmosphere) further south that at higher U (higher in the atmosphere). So, implying a northerly tilt with height of our vortex tower!
